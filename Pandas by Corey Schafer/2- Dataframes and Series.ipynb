# Dataframes and Series

## Dataframes
we might see defintion of dataframe, 2-dimensional datastructure  
but it is basically a table with rows and columns   
columns are keys of dictionaries and   
rows are values in the list of keys

```python
import pandas as pd
```

```python
#  dummy dictionary to convert it into dataframe

data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [25, 32, 18, 47],
    'shape': ['New York', 'Paris', 'London', 'San Francisco']
}
```

```python
# converting into dataframe 

df= pd.DataFrame(data)
```

```python
df

# 0,1,2,3 are indexes, this a column added by the pandas
# we can create our own index column, will see later
# index column generated by the pandas is unique but
# we can set any column as index column if it has unique values of not
```

## Checking type of dataframe

```python
type(df)
```

## Series

we might see definition online, like 1-dimensional data structure 
but series are just the rows of the dataframe  
it is just a list with additional functionalities of pandas  
#### We can think it, a dataframe contains multiple object of series

```python
df["name"]
```

```python
# if we check the type of above line, it is series 

type(df["name"])
```
