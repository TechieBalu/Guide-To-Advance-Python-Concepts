# Dataframes and Series

## Dataframes
we might see defintion of dataframe, 2-dimensional datastructure  
but it is basically a table with rows and columns   
columns are keys of dictionaries and   
rows are values in the list of keys

```python
import pandas as pd
```

```python
#  dummy dictionary to convert it into dataframe

data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [25, 32, 18, 47],
    'shape': ['New York', 'Paris', 'London', 'San Francisco']
}
```

```python
# converting into dataframe 

df= pd.DataFrame(data)
```

```python
df

# 0,1,2,3 are indexes, this a column added by the pandas
# we can create our own index column, will see later
# index column generated by the pandas is unique but
# we can set any column as index column if it has unique values of not
```

## Checking type of dataframe

```python
type(df)
```

## Series

we might see definition online, like 1-dimensional data structure 
but series are just the rows of the dataframe  
it is just a list with additional functionalities of pandas  
#### We can think it, a dataframe contains multiple object of series

```python
df["name"]
```

```python
# if we check the type of above line, it is series 

type(df["name"])
```

#### Dot notation to get the series out of the dataframe

Dot notaion and bracket notation gives the same results

bracket notation is preferred because, there is a chance that our column can be named as the **dataframe's** attribute or function e.g. **shape**, shape can be our column and shape is also the attribute of the **dataframe** 

if we have a column name same as the attribute or function, Pandas will give preference to the attribute or function, So, bracket notation is more useful

```python
df.name
```

```python
# printing "shape" column which is also the attribute of the dataframe, 
# pandas gives results on the attribute not the column name
df.shape
```

## Accessing multiple column of dataframe

we use the double brackets to access multiple columns and  
resultant is not **series** datatype rather it will be **dataframe** datatype, we can think of it a subset of dataframe  

```python
# shape is city basically
df[["name","shape"]]
```

```python
# type of above line
type(df[["name","shape"]])
```

```python
# We can also access all the columns name by using the columns attribute
df.columns
```

```python
# type of above line will give, index class object
type(df.columns)
```
