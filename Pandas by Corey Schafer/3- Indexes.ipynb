## Setting up custom index column
Most of the cases, pandas does not restricts to have a custom index column which have duplicated values but but it might be the case where pandas restricts to set custom index column without any duplicate value.  
to set the index we use **set_index()** function  
resetting index: **reset_index()**

```python
# setting "Respondent" column as indexed column 
df.set_index("Respondent")

# as you can see, "Respondent" column is now below from all other columns in alignment, it is set as index column 
# let's discuss the benefits of setting a column as index
```

```python
# if we print our original dataset, we can see pandas did not change the dataset,
df

# still index column in present, but if we want to set the index right on the main dataset, we set "inplace" argument as true
```

```python
df.set_index("Country", inplace=True)
```

```python
df

# Now, if I print the dataframe, "Country" column in set as index on the original dataset
```

```python
# we can print the index column just by using the "index" attribute of pandas

df.index
```
