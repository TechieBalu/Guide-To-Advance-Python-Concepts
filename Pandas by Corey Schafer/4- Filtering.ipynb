## Filtering

Filtering is all about getting the specific data from columns and apply multiple gate operations on it like OR, AND, NOT

```python
# Let's filter the records, where "Country" column has value "Pakistan"
df["Country"] == "Pakistan"

# This is giving us all the rows (Series) and shows results in boolean form, either "Country" column has value "Pakistan" or not
# We can convert them into actual dataframes
```

```python
results = df["Country"] == "Pakistan"
df[results]

# this is giving us all those rows that has value "Pakistan" in "Country" column
```

```python
# Short hand notation of filtering: 

df[df["Country"] == "Pakistan"]
```
